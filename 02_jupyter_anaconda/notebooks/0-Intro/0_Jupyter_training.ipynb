## How Do Jupter Notebooks Help Perform & Communicate Science?

Jupyter Notebooks help perform and communicate science in many phases:

- Experimental design phase
- Grant writing phase
- Analysis phase
- Collaboration phase
- Publication phase

### Experimental Design Phase:

- Beging writing narrative that will drive:
    - Grant application
    - Actual research 
    - A possible scientifi paper
- Assist in preliminary data analysis
- Keeps a record of all thoughts and steps in the process

### Grant Writing Phase:

- Take record of what was written
- Clean it up
- Use output from preliminary data analysis in application
- Expand analysis if resubmitting grant

### Grant received... :-)

### Analysis Phase:

- Jupyter notebook are a record of the though process and exploratory workflow
- Using existing Notebook to communicate project to PostDoc or grad student who will do the work
- Enables researcher to see preliminary data analysis and teh narative that surronds it
- Combine existing tools and scripts into a single workflow

- NOTE: I start by using a single notebook for my workflow, creating a new cell for each change or iteration of my analysis. Only after I am satisfied with the results of my analysis, do I duplicate my notebook and CAREFULLY delete cells, which are not part of the final workflow.

### Collaboration Phase:

- Sending a Jupyter notebook as a single document to a collaborator or to PI
- They can rerun analysis, edit, remove or add to it (if they have teh software installed)

### Communication Phase

- Use the notebook as the start of a manuscript
- Examples
    - Reproducible Science Curriculum - [Publication](https://reproducible-science-curriculum.github.io/publication-RR-Jupyter/aio.html)
    - [Jupyter notebook manuscript workflow](https://sylvaindeville.net/2015/07/17/writing-academic-papers-in-plain-text-with-markdown-and-jupyter-notebook/)
- Sumbit notebook as suplementsry material
- Future
    - Submit notebook as actual manuscript?
